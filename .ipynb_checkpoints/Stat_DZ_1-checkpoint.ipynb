```python
import pandas as pd
df = pd.read_csv('horse_data.csv', header=None, na_values = '?')
df8 = df[[(0), (1), (3), (4), (5), (6), (10), (22)]]
df8.columns = ['Surgery', 'Age', 'rectal temperature', 'pulse', 'respiratory rate', 'temperature of extremities', 'pain', 'outcome']
df8
```

##### 1:  surgery
          1 = Yes, it had surgery
          2 = It was treated without surgery
##### 2:  Age
          1 = Adult horse
          2 = Young (< 6 months)
##### 3:  rectal temperature
          - linear
          - in degrees celsius.
          - An elevated temp may occur due to infection.
          - temperature may be reduced when the animal is in late shock
          - normal temp is 37.8
          - this parameter will usually change as the problem progresses
               eg. may start out normal, then become elevated because of
                   the lesion, passing back through the normal range as the
                   horse goes into shock
##### 4:  pulse
          - linear
          - the heart rate in beats per minute
          - is a reflection of the heart condition: 30 -40 is normal for adults
          - rare to have a lower than normal rate although athletic horses
            may have a rate of 20-25
          - animals with painful lesions or suffering from circulatory shock
            may have an elevated heart rate
##### 5:  respiratory rate
          - linear
          - normal rate is 8 to 10
          - usefulness is doubtful due to the great fluctuations
##### 6:  temperature of extremities
          - a subjective indication of peripheral circulation
          - possible values:
               1 = Normal
               2 = Warm
               3 = Cool
               4 = Cold
          - cool to cold extremities indicate possible shock
          - hot extremities should correlate with an elevated rectal temp.
##### 7: pain 
          - a subjective judgement of the horse's pain level
          - possible values:
               1 = alert, no pain
               2 = depressed
               3 = intermittent mild pain
               4 = intermittent severe pain
               5 = continuous severe pain
          - should NOT be treated as a ordered or discrete variable!
          - In general, the more painful, the more likely it is to require
            surgery
          - prior treatment of pain may mask the pain level to some extent
##### 8: outcome
          - what eventually happened to the horse?
          - possible values:
               1 = lived
               2 = died
               3 = was euthanized

Age  - пропусков нет
 
 Surgery , outcome - пропуск один в одной и тойже строчке, судя по остальным показателем скорее всего данной особи могла бы
быть проведена операция, но данных по результату операции тоже нет, и каких либо выводов по успешности операции мы получить не
можем так что эту строчку я бы просто удалил

rectal temperature - по данным полученным в прошлом задании можно сказать в общем все температура имеет минимальный разброс так что данный пункт можно заполнить медианным значением
    
temperature of extremities - в описании данных сказано что данная температура имеет взаимосвязь с rectal temperature, используем медианное значение на основании rectal temperature 

respiratory rate - в описании сказано что что это сомнительные показатели, 
так как имеют слишком большой разброс значений, их так же можно заполнить медианным показателем 
    
pain - как указано в описании высокий показатель боли напрямую ведет к операции заполняем медианным значением на основании Surgery        
 
pulse - опять же по описанию зависит от наличия боли заполняем медианным показателем на основании pain

```python
ndf = df8.drop(index=[132])
ndf['rectal temperature'].fillna(ndf['rectal temperature'].median(), inplace=True)
ndf['temperature of extremities'].fillna(ndf.groupby(['rectal temperature'])['temperature of extremities'].transform('median'), inplace=True)
ndf['respiratory rate'].fillna(ndf['respiratory rate'].median(), inplace=True)
ndf['pain'].fillna(ndf.groupby(['Surgery'])['pain'].transform('median'), inplace=True)
ndf['pulse'].fillna(ndf.groupby(['pain'])['pulse'].transform('median'), inplace=True)
```
