```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
gogl = pd.read_csv('googleplaystore.csv')
gogl.head()
```

Соберем статистику по самым популярным приложениям в магазине Гугла, для этого отфильтруем дата сет по количеству установок(1,000,000,000+), а так же удалим дубликаты так как строки по некоторым приложениям повторяются.

```python
topgogl=gogl[gogl['Installs']=='1,000,000,000+']
gp=topgogl.drop_duplicates(subset=['App'], keep='first')
gp
```

```python
gp.groupby('Category')['App'].count().sort_values(ascending=False).plot(kind='bar',
                                                                                xlabel='Категория',
                                                                                ylabel='Кол-во приложений',
                                                                                title = 'Категории самых популярных приложений')
```

По столбчатой диаграмме видим что основное количество топовых приложений приходится на Приложение для связи и соцсети.

```python
gp['Rating'].plot(kind = 'box', title = 'Разброс оценок топовых приложений')
```

По "Ящику с усам" мы видим что средняя оценка у приложений 4,3. Большинство оценок лежит в диапозоне 4,1-4,4. И что имеются выбросы с минимальным количеством оценок в 4,5 и 3,7

```python
cnt_r=gp.groupby('Content Rating')['App'].count()
cnt_r.plot(kind='pie',
                   title='Распределение возрасного рейтинга')
```

По круговой диаграмме видно что большая часть приложений доступны для всех возрастов. Но разница минималь с приложениями имеющими возрастной рейтинг.

```python
topgogl.groupby('Content Rating')['App'].count().sort_values(ascending=True).plot(kind='barh', 
              title='Распределение по рейтингу')
```

Более наглядное распределение количесва приложений по рейтингу
